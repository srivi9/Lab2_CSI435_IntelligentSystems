```python
#these will import the libraries we need
import matplotlib.pyplot as plt
import numpy as np
```

```python
# What does this function do?
#This function finds the gradient and multiplies it by
#the multiplier, then add it to the value for x. This moves our current location in the direction of
# the steepest ascent i.e. up the slope
def GradientAscent1(a):

    grad = -2.0*a
    mult = 0.05
    mult_grad=mult*grad
    a = a + mult_grad

    # complete this
    return a
```

```python
# What does this function do?
# this function finds the gradient and multiplies it by
# the multiplier, then adds it to the value for x.This moves the current location in the direction of
# the steepest ascent i.e. up the slope
def GradientAscent2(a):
    mult=0.05
    grad = (np.exp(-a*a)*(6.0*np.sin(3*np.pi*a)*np.cos(3*np.pi*a)) - 2.0*a*np.sin(3.0*np.pi*a)*np.exp(-a*a))
    mult_grad=mult*grad
    a = a + mult_grad
    # complete this
    return a
```

```python
# Gradient descent function for Obj3 (Minimizing)
# this function finds the gradient and multiplies it by
# the multiplier, then subtracts it from the value for x.This moves the current location in the direction of
# the steepest descent i.e. down the slope
def GradientDescent3(a):
    grad = 2 * a  # the Derivative of x^2 is 2x
    mult = -0.00002   # Step size
    a = a - mult * grad
    return a
```

```python
# Gradient ascent function for Obj4 (Maximizing)
# this function finds the gradient and multiplies it by
# the multiplier, then adds it to the value for x.This moves the current location in the direction of
# the steepest ascent i.e. up the slope
def GradientAscent4(a):
    grad = 6*a**2 + 8*a  # Derivative of 2x^3 + 4x^2 + 1 is 6x^2 + 8x
    mult = 0.023       # Step size
    a = a + mult * grad
    return a
```

```python
#these are all the functions of the curves to be drawn.
def Obj1(x):
    return -x*x

def Obj2(x):
    return np.exp(-x*x)*(np.sin(3.0*np.pi*x)**2)

def Obj3(x):
    return x*x

def Obj4(x):
    return 2*x**3+4*x**2+1
```

```python
#this is the hillclimb algorithm that is used to find the local maximum
def HillClimb(x,next_func,objective_func,fig):
    current = x #stores the current state
    max_iterations = 0 #to prevent an infinite loop
    while max_iterations < 10000:
      successors = next_func(current) #this will get the successors

      if objective_func(successors) < objective_func(current): #if the current state better than successors
        return current                                            #then return the current state

      plt.plot(successors, objective_func(successors), 'ro')    #plot the successors

      current = successors   #updates the current state
      max_iterations += 1    #increment the iterations by one

    # hint: printing the progress
    print (current, objective_func(current), objective_func(successors))



```

```python
#generating a range of numbers for our x values
b = np.arange(-3, 3, 0.02)

# plot the two functions here, in separate figures
plt.figure(1)
plt.plot(b, Obj1(b))

# display a title
plt.title("y = -x * x")


#ploting the functions
plt.figure(2)
plt.plot(b, Obj2(b))

# display a title
plt.title("y = exp(-x * x) * sin(3 * Pi * x)^1")

#this is our start location
start=-1.6

# call HillClimbing for the two functions to be optimized
HillClimb(start, GradientAscent1, Obj1, plt.figure(1))

HillClimb(start, GradientAscent2, Obj2, plt.figure(2))



```

```python


# For Obj3 (this is Minimization)
start_location3 = -1.6  #a starting point where function is decreasing to ensure convergence

#this is to plot the figures
fig3 = plt.figure(1)
plt.plot(b,Obj3(b))

#this is calling hill climb for obj3
HillClimb(start_location3, GradientDescent3, Obj3, fig3)

# For Obj4 (this is Maximization)
start_location4 = -3.0  #a starting point where the function is increasing to ensure convergence

#this is to plot the figures
fig4 = plt.figure(2)
plt.plot(b,Obj4(b))

#this is calling hill climb for obj4
HillClimb(start_location4, GradientAscent4, Obj4, fig4)

```
